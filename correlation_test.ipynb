```python
import numpy as np

x = np.arange(0, 50)
print(x)
```

สร้างเลขสุ่ม

```python
import random as rd
y = rd.randrange(1, 100, 5)
print(y)
```

หาเลขสุ่มจำนวน 50 ค่า

```python
y = []

for i in range(0, 50):
  a = rd.randrange(1, 100)
  y.append(a)
```

สร้างกราฟ scatter จากค่า x และ y (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
```

หาค่าความสัมพันธ์ (correlation)

```python
corr = np.corrcoef(x,y)
print(corr)
```

หาค่า linear regression (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html)

```python
import scipy.stats

slope, intercept, r, p, stderr = scipy.stats.linregress(x, y)
print(slope, intercept, r, p, stderr)
```

สร้างกราฟ

```python

#create chart
print("correlation: "+ str(corr[0][1]))
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=0, marker='o', label="data")
ax.plot(x, intercept + slope * x, label="fit line")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(facecolor='white')
ax.set_facecolor('#dbdbdb')
ax.grid(color='white', linestyle='solid')
plt.show()
```
